```python
import os
import zarr
import glob
import xarray as xr
from rechunker import rechunk
from dask.diagnostics import ProgressBar
from dask.distributed import Client, LocalCluster
```

```python
ensmems = ["01", "04", "06", "15"]
varnames = [
    "tmean",
    "tmin",
    "tmax",
    "hurs",
    "huss",
    "psurf",
    "rlds",
    "rsds",
    "sfcWind",
    "pr",
]

for ensmem in ensmems:
    for varname in varnames:
        if varname == "psurf" and not ensmem == "01":
            continue
        else:
            print("Processing ensmem " + ensmem + " " + varname)
            INPUT = (
                "/gws/nopw/j04/ceh_generic/matbro/chess_scape/monthchunk/ens"
                + str(ensmem)
                + "-monthchunk/"
                + varname
                + ".zarr"
            )
            OUTPUT_DIRECTORY = (
                "/gws/nopw/j04/ceh_generic/matbro/chess_scape/monthchunk2/ens"
                + str(ensmem)
                + "-monthchunk"
            )
            OUTFILE = os.path.join(OUTPUT_DIRECTORY, varname + ".zarr")
            if not os.path.exists(OUTPUT_DIRECTORY):
                os.makedirs(OUTPUT_DIRECTORY)

            origzarr = xr.open_zarr(INPUT)
            origzarr["lat"] = origzarr["lat"][0, :, :]
            origzarr["lon"] = origzarr["lon"][0, :, :]

            with ProgressBar():
                origzarr.to_zarr(OUTFILE)
```
